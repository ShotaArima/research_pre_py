7.課題:コッホ曲線

```python
#課題　コッホ曲線の描画
from math import sqrt#mathライブラリの平方根sqrt
from PIL import Image,ImageDraw#描画に必要なライブラリ
```

```python
def length(a):#長さの自動調節を行う関数length
  x,y = 0,0
  for (dx,dy) in a:
    x += dx
    y += dy
  return sqrt(x**2 + y**2)
```

```python
#確認用①
a = [(1,0),(0,1)]
length(a)
```

```python
def convert(a,b):
  ax,ay = a
  alen =sqrt(ax**2+ay**2)
  c=ax/alen
  s=ay/alen
  scale = alen/length(b)

  #リスト内包表記不使用の場合
  r=[]
  for(bx,by) in b:
    bx*=scale        #(a)
    by*=scale        #(a)
    nx=c*bx-s*by     #(b)
    ny=s*bx+c*by     #(b)
    r.append((nx,ny))
  return r

  #Q1.リスト内包表記にするとどうなる？
  #b=[○○○○ for (x,y) in b]   #(a)
  #b=[○○○○ for (x,y) in b]　 #(b)
  #return b
```

```python
#確認用②
a=(0,1)
b=[(1,1),(1,-1)]
convert(a,b)
```

```python
def apply(a,b):
  r=[]
  for i in a:
    r += convert(i,b)
  return r
```

```python
#確認用③
a=[(1,0),(0,-1)]
b=[(1,1),(1,-1)]
apply(a,b)
```

```python
def draw_line(draw,a,size):
  x1,y1=0,0
  for(dx,dy) in a:
    x2=x1+dx
    y2=y1+dy
    draw.line((x1,size/2-y1,x2,size/2-y2),fill=(255,255,255))
    x1,y1=x2,y2
```
