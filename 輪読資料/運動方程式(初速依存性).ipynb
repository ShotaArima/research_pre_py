# 初速依存性のある運動方程式

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
from math import pi, cos, sin
from pylab import rcParams
```

```python
# 初速2の場合の運動方程式の数値解法
def throw2(theta, y0):
  rx, ry=0.0, y0
  vx, vy=2.0*cos(theta), 2.0*sin(theta) # 書き換え箇所
  ax, ay=[], []
  g=1.0
  h=0.001
  while ry>=0.0:
    rx+=vx*h
    ry+=vy*h
    vy-=g*h
    ax.append(rx)
    ay.append(ry)
  return ax, ay
```

```python
# プロット2
def plot(angles, y0=0.0):
  for theta in angles:
    nx, ny=throw2(theta/180.0*pi, y0)
    plt.plot(nx, ny, label=theta)
  plt.legend()
  plt.show()
```

```python
# 地面から投げた場合
angles=[40, 45, 50]
plot(angles)
```
