# 運動方程式の実践

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
from math import pi, cos, sin
from pylab import rcParams
```

```python
# プロットサイズの変更
rcParams['figure.figsize']=10, 5
```

```python
# 運動方程式の数値解法
def throw(theta, y0):
  rx, ry=0.0, y0
  vx, vy=cos(theta), sin(theta)
  ax, ay=[], []
  g=1.0
  h=0.001
  while ry>=0.0:
    rx+=vx*h
    ry+=vy*h
    vy-=g*h
    ax.append(rx)
    ay.append(ry)
  return ax, ay
```

```python
# プロット
def plot(angles, y0=0.0):
  for theta in angles:
    nx, ny=throw(theta/180.0*pi, y0)
    plt.plot(nx, ny, label=theta)
  plt.legend()
  plt.show()
```

```python
# 地面から投げた場合
angles=[40, 45, 50]
plot(angles)
```
