```python
# モンテカルロ法による数値分析
from random import random

def calc_pi(n):
  r=0
  for _ in range(n):
    x=random()
    y=random()
    if x**2 + y**2 <1.0:
      r+=1
  return 4*r/n

calc_pi(10000)
```

```python
# モンティ・ホール問題(keep派)
from random import choice, seed
from copy import copy
```

```python
def keep(boxes):
  answer=choice(boxes)
  first_choice=choice(boxes)
  return answer==first_choice
```

```python
seed(1)
boxes=[1,2,3]
for _ in range(10):
  print(keep(boxes))
```

```python
# モンティ・ホール問題(change派)
def change(boxes):
  answer=choice(boxes)
  first_choice=choice(boxes)
  rest_boxes=boxes.copy()
  rest_boxes.remove(answer)
  if first_choice==answer:
    second_choice=choice(rest_boxes)
  else:
    second_choice=answer
  return answer==second_choice
```

```python
seed(1)
boxes=[1,2,3]
for _ in range(10):
  print(change(boxes))
```

```python
# 確率の計算
seed(1)
boxes=[1,2,3]
k=0
c=0
n=10000
for _ in range(n):
  if keep(boxes):
    k+=1
  if change(boxes):
    c+=1
print("keep:"+str(k/n))
print("change"+str(c/n))
```

```python
def find(i,parent):
  while i !=parent[i]:
    i=parent[i]
  return i
```

```python
def union(i, j, parent):
  i=find(i, parent)
  j=find(j, parent)
  parent[j]=i
```

```python
def make_conf(L,p):
  parent=[i for i in range(L*L)]
  for iy in range(L-1):
    for ix in range(L-1):
      i=ix+iy*L
      j=ix+1+iy*L
      if random.random()<p:
        union(i, j, parent)
      j=ix+(iy+1)*L
      if random.random()<p:
        union(i, j, parent)
  return parent
```

```python
def gacha(n):
  cd=0
  posters=[]
  while len(set(posters))<n:
    posters.append(random.randint(1,n))
    cd+=1
  return cd
```
