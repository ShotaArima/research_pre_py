## 1,文字とは何か  
  - 字体(glyph)...ある文字の持つ共通的な特徴  
    ↓  
  - 文字コード(character code)...字体に対して何らかのユニークな番号を振るコード  
    ↓  
  - 符号化(encodng)...文字コードをメモリ上での格納するために行う  
    ↓  
  - 書体(font)...メモリに格納された文字をディスプレイに表示させる際の文字の形  



## 2,辞書
  ### 復習  
  リストは複数の要素をまとめるデータ構造。「要素」と添え字の「数字」を結びつけるデータ構造。インデックスによって要素にアクセスすることができる。

---  

  ### ハッシュ
   「数字以外の何か」と「要素」を結びつけるデータ構造  
  ### 辞書
   Pythonでのハッシュを指す  
  ### キー
   要素にアクセスするためのもの





````python
#辞書型は中かっこ{}で初期化する。
d={}
````

````python
d["Apple"]=158
print(d["Apple"])
````

````python
#「キー:値」という形で初期化時に要素を指定できる
d={"Apple":158,"Banana":198,"Orange":100}
````

````python
#なんでもキーにすることができる
d[1]="one"
d[(1,2)]=(2,4)
````

````python
for k in d:
  print(f"k={k}")
  print(f"d[k]={d[k]}")
````

````python
#キーと要素を同時に取得する場合
for k,v in d.items():
  print(k,v)
````

````python
#存在しないキーの指定
d["new"]
#エラー:KeyError: 'new'
````

````python
#存在しないキーが指定した時にデフォルトの値があると便利な時がある
from collections import defaultdict
````

````python
import collections
#デフォルトの値として「0」を持つような辞書は以下の通り
d=defaultdict(int)
#出現するものの頻度をカウントすることができる
````

````python
from collections import defaultdict
d=defaultdict(int)
s="すもももももももものうち"
for c in list(s):
  d[c]+=1
for k,v in d.items():
  print(k,v)
````

text='隴西《ろうさい》の李徴《りちょう》は博学|才穎《さいえい》、天宝の末年、若くして名を虎榜《こぼう》に連ね、ついで江南尉《こうなんい》に補せられたが、性、狷介《けんかい》、自《みずか》ら恃《たの》むところ頗《すこぶ》る厚く、賤吏《せんり》に甘んずるを潔《いさぎよし》としなかった。'  

---

- 文字を一文字ずつ処理
- 《》に囲まれた状態であるかを判定
  - 囲まれた状態ならスキップ
  - 囲まれていないなら表示  

といったアルゴリズムになる






````python
text='隴西《ろうさい》の李徴《りちょう》は博学|才穎《さいえい》、天宝の末年、若くして名を虎榜《こぼう》に連ね、ついで江南尉《こうなんい》に補せられたが、性、狷介《けんかい》、自《みずか》ら恃《たの》むところ頗《すこぶ》る厚く、賤吏《せんり》に甘んずるを潔《いさぎよし》としなかった。'
````

````python
in_bracket=False
for s in list(text):
  if in_bracket:
    if s=='》':
      in_bracket=False
    continue
  if s=='《':
    in_bracket=True
    continue
  print(s,end="")
print()
````

````python
#正規表現を使用すると以下のように表現できる
import re
print(re.sub(r'《.*?》','',text))
````

呪文のような文字列「^#」や「.*?」が正規表現である

2. 青空文庫からのデータ取得

````python
#ライブラリのインポート
from collections import defaultdict
import re
import io
import urllib.request
from zipfile import ZipFile
````

````python
#Webからデータを取得する関数
def load_from_url(url):
  data=urllib.request.urlopen(url).read()
  zipdata=ZipFile(io.BytesIO(data))
  filename=zipdata.namelist()[0]
  text=zipdata.read(filename).decode("shift-jis")
  text=re.sub(r'[.*?]','',text)
  text=re.sub(r'《.*?》','',text)
  return text
````

````python
#青空文庫からのデータの取得テスト
URL="https://www.aozora.gr.jp/cards/000119/files/624_ruby_5668.zip"
text=load_from_url(URL)
text.split()[0]
````
